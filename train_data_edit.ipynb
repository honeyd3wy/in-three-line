```python
import pandas as pd

json = pd.read_json('train_original.json')
docs = json[['documents']]

docs
```

## article 추출

```python
article_lst = []
for i in range(len(docs)):
    doc = docs.iloc[i][0]['text']
    text = list(itertools.chain(*doc))

    article = [t['sentence'] for t in text]
    article_lst.append(" ".join(article))

    # for t in text:
    #     print(t['sentence'])

article_lst
```

```python
len(article_lst)
```

## abstractive 추출

```python
docs.iloc[1][0]['abstractive'][0]
```

```python
abstractive_lst = [docs.iloc[i][0]['abstractive'][0] for i in range(len(docs))]
abstractive_lst
```

```python
len(abstractive_lst)
```

## 데이터 합치기

```python
train_set1 = pd.DataFrame({'news':article_lst,
                            'summary':abstractive_lst})
train_set1.head()
```
